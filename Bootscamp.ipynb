```python
#Question 7

#1. select all the ´Lotus" cars from the data set

lotus=df_first[df_first['Make']=='Lotus']
lotus
print("-------------------------------------------------------")
#2. select only columns " Engine HP", "Engine Cylinders"
two_columns=lotus[['Engine HP', 'Engine Cylinders']]
two_columns
#Now drop all duplicated rows using drop_duplicate
drop=two_columns.drop_duplicates()

#now getting the arrays
x=drop.to_numpy()
x

#Compute matrix-matrix multiplication between the transpose of X and X. To get the transpose, use X.T. Let's call the result XTX.
xtx=np.dot(x.T, x)
xtx

#Invert XTX.
xtx_inv=np.linalg.inv(xtx)
xtx_inv

#Create an array y

y=np.array([1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800])

#Create w
w=np.dot(np.dot(xtx_inv, x.T), y)
w
```
